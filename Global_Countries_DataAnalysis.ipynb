# Loading Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
country_data = pd.read_csv('world-data-2023.csv')
```

```python
country_data.head()
```

```python
country_data.columns
```

# Dropping Columns

```python
country_data.drop(columns=['Abbreviation','Calling Code','Capital/Major City','Currency-Code','Largest city',
                            'Official language','Latitude','Longitude'], axis=1, inplace=True)
```

```python
country_data.columns
```

```python
country_data.info()
```

# Data Cleaning

```python
country_data.dtypes
```

### Changing data type of float

```python
# changing data type to float
columns_to_change = ['Density\n(P/Km2)', 'Agricultural Land( %)', 'Land Area(Km2)','Co2-Emissions', 'Forested Area (%)',
                      'CPI', 'CPI Change (%)', 'Gasoline Price', 'GDP','Gross primary education enrollment (%)', 'Armed Forces size',
                      'Gross tertiary education enrollment (%)','Minimum wage','Out of pocket health expenditure','Population',
                      'Population: Labor force participation (%)','Tax revenue (%)', 'Total tax rate', 'Unemployment rate', 'Urban_population']
country_data[columns_to_change] = country_data[columns_to_change].map(
        lambda col: float(str(col).replace(',','').replace('%','').replace('$','')))
```

```python
country_data.dtypes
```

```python
country_data.head()
```

### Checking null/missing values

```python
country_data.isnull().sum()
```

```python
# Removing rows which contains null in almost all columns
country_data.dropna(thresh = 22, axis=0, inplace=True)
```

```python
# Replacing null with 0 in armed forces size column
country_data.fillna(value={"Armed Forces size":0}, inplace=True)
```

```python
# Replacing null with mode in remaining columns
columns_with_mode = ['Gasoline Price','CPI','CPI Change (%)','Gross primary education enrollment (%)','Gross tertiary education enrollment (%)',
                    'Maternal mortality ratio','Minimum wage','Out of pocket health expenditure','Population: Labor force participation (%)',
                    'Tax revenue (%)','Total tax rate','Unemployment rate']
country_data[columns_with_mode] = country_data[columns_with_mode].fillna(country_data[columns_with_mode].mode().iloc[0])
```

```python
# Checking for any null value
country_data.isnull().sum()
```

```python
country_data.info()
```

# Feature Engineering

```python
# Adding relevant features for data analysis and visualization
# eg. GDP per capita, CO2-emission per capita etc.

country_data['GDP per capita'] = country_data['GDP']/country_data['Population']
country_data['CO2-Emission per capita'] = country_data['Co2-Emissions']/country_data['Population']
country_data['Armed Forces size per capita'] = country_data['Armed Forces size']/country_data['Population']
country_data['Agricultural Land(Km2)']=country_data['Agricultural Land( %)']*country_data['Land Area(Km2)']/100
country_data['Forested Area(Km2)']=country_data['Forested Area (%)']*country_data['Land Area(Km2)']/100
country_data['Urban_population ratio']=country_data['Urban_population']/country_data['Population']
```

### Top 10 countries by GDP & GDP per capita

```python
# sorting by gdp per capita in descending order
top_country_by_gdp = country_data[['Country','GDP']].sort_values(by='GDP',ascending=False).head(10)
ax = top_country_by_gdp.set_index("Country").plot.bar(color="#C40D07",legend=False)

plt.ylabel("GDP")
plt.xlabel("Country")
plt.title("Top 10 Countries by GDP")
plt.show()

top_country_by_gdp_perCapita = country_data[['Country','GDP per capita']].sort_values(by='GDP per capita',ascending=False).head(10)
ax = top_country_by_gdp_perCapita.set_index("Country").plot.bar(color="#0818A1",legend=False)
ax.bar_label(ax.containers[0],color="black",fontsize=8)

plt.ylabel("GDP per Capita")
plt.xlabel("Country")
plt.title("Top 10 Countries by GDP per capita")
plt.show()
```
